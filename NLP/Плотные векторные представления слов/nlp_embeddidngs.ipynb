### Плотное векторное представление Navec для отзывов на банки

```python
!pip install pymorphy3
```

```python
import pandas as pd
import pymorphy3
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def preprocess(text, stop_words, punctuation_marks, morph):
    tokens = word_tokenize(text.lower())
    preprocessed_text = []
    for token in tokens:
        if token not in punctuation_marks:
            lemma = morph.parse(token)[0].normal_form
            if lemma not in stop_words:
                preprocessed_text.append(lemma)
    return preprocessed_text
```

```python
punctuation_marks = ['!', ',', '(', ')', ':', '-', '?', '.', '..', '...']
stop_words = stopwords.words("russian")
morph = pymorphy3.MorphAnalyzer()
```

```python
!wget https://www.dropbox.com/s/a9r0b2yj3vqvi13/banks.csv?dl=1 -O banks.csv
```

```python
banks = pd.read_csv('banks.csv', sep='\t', index_col='idx');
```

```python
banks
```

```python
text = banks.iloc[0]['Text']
```

```python
preprocessed_text = preprocess(text, stop_words, punctuation_marks, morph)
```

```python
preprocessed_text
```

Векторные представления слов Navec для отзыва о банке

```python
navec['альфа-банк']
```

```python
navec['работать']
```

Находим плотные векторные представления для всех слов в тексте

```python
text_embeddings = []
for tocken in preprocessed_text:
    embedding = navec[tocken]
    text_embeddings.append(embedding)
```

**Учитываем неизвестные слова**

Вектор для неизвестного слова в Navec

```python
navec['<unk>']
```

```python
unk = navec['<unk>']
text_embeddings = []
for tocken in preprocessed_text:
    embedding = navec.get(tocken, unk)
    text_embeddings.append(embedding)
```

Векторное представление для текста

```python
text_embeddings
```

Длина векторного представления текста

```python
len(text_embeddings)
```
